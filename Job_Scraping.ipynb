```python
import csv
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def extract(jobs,page):
  headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36"}
  url=f'https://merojob.com/search/?q={jobs}&page={page}'
  response=requests.get(url,headers)
  soup=BeautifulSoup(response.text,'html.parser')
  return soup
```

```python

def value(soup):
  tech_list=[]
  cards=soup.find_all('div',class_='card mt-3 hover-shadow')
  for items in cards:
    listed=[]
    Job=items.find('a').text.strip()
    Office=items.find('h3').text.strip()
    deadline=items.find('p').text.strip().strip('Apply Before:')
    location=items.find('div',class_='media-body').text.strip()
    tool=items.find_all('span',class_='badge badge-pill badge-light rounded text-muted')
    for technology in tool:
      tech_list.append(technology.text.strip())
    values={
        'Office':Office,
        'Position':Job,
        'Location':location,
        'Deadline':deadline,
        'Skills':tech_list,
    }
    job_list.append(values)
    tech_list=[]
  return
```

```python
jobs=input("Enter the position that you want to see the vacancy of:")
job_list=[]
range_index=int(input("Enter the number of pages:"))
if jobs.isalpha():
    # Your code to run the problem goes here
     for i in range(1,range_index,1):
      c=extract(jobs,i)
      value(c)
```
